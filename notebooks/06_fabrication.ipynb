# 1. Process Overview
---

*Content to be developed*

Typical SCE fabrication flow:
1. Substrate preparation (Si or sapphire)
2. Ground plane deposition
3. Insulator layers
4. Trilayer deposition and junction definition
5. Wiring layers
6. Resistor layers (if needed)
7. Passivation and pad opening

*Content to be developed*

<a id="3-lithography"></a>
# 3. Lithography
---

*Content to be developed*

| Technique | Resolution | Use |
|-----------|------------|-----|
| Contact/proximity | ~1 µm | Coarse features |
| Stepper (i-line) | ~0.5 µm | Standard production |
| DUV stepper | ~0.25 µm | Advanced nodes |
| E-beam | <100 nm | R&D, small junctions |

SCE feature sizes typically 0.25-2 µm (less demanding than CMOS)

---
<a id="4-etching"></a>
# 4. Etching
---

*Content to be developed*

- **RIE (Reactive Ion Etching)**: Anisotropic, common for Nb
- **ICP-RIE**: Higher density plasma, smoother sidewalls
- **Wet etching**: Isotropic, used for some dielectrics
- **Ion milling**: Physical sputtering, non-selective

Nb typically etched with SF6/O2 or CF4/O2 chemistry

---
<a id="5-junction-formation"></a>
# 5. Josephson Junction Formation
---

*Content to be developed*

**Nb/Al-AlOx/Nb trilayer process:**

1. Deposit Nb base electrode (~150-200 nm)
2. Deposit Al wetting layer (~7 nm) in-situ
3. Oxidize in controlled O2/Ar atmosphere
4. Deposit Nb counter electrode (~100 nm)
5. Pattern junction area (self-aligned or defined)
6. Anodize or etch counter electrode

Critical current density J_c controlled by oxidation (time, pressure, temperature)

<a id="6-layer-stacks"></a>
# 6. Layer Stacks
---

*Content to be developed*

Typical SFQ process (e.g., MIT-LL SFQ5ee):

| Layer | Material | Thickness | Purpose |
|-------|----------|-----------|--------|
| M0 | Nb | 200 nm | Ground plane |
| I0 | SiO2 | 150 nm | Insulator |
| M1 | Nb | 200 nm | Wiring |
| I1 | SiO2 | 100 nm | Insulator |
| Trilayer | Nb/Al/Nb | ~300 nm | Junctions |
| I2 | SiO2 | 200 nm | Insulator |
| M2 | Nb | 300 nm | Top wiring |
| R | Mo or PdAu | 50 nm | Resistors |

# 7. Process Control and Characterization
---

*Content to be developed*

| Measurement | Technique | Purpose |
|-------------|-----------|--------|
| Film thickness | Ellipsometry, profilometry | Process control |
| Sheet resistance | 4-point probe | Film quality |

| T_c | 4-point at cryo | Material quality |

## Summary

- SCE fabrication uses **standard semiconductor techniques**
- **Nb trilayer** is the standard junction technology
- Feature sizes are **relaxed** compared to modern CMOS
- **J_c control** is critical for circuit yield
